```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df_news_7y = pd.read_csv('news_20151114_drop_prep.csv')
df_news_10y = pd.read_csv('news_20080401.csv')
df_news_1y = pd.read_csv('news_20180701.csv')
```

```python
df_retry2 = pd.read_csv('retry_url_080401~151115.csv')
df_retry1 = pd.read_csv('retry_url_151115~221115.csv')
```

```python
df_retry2.fillna('', inplace=True)
df_retry2
```

```python
df_retry2['date_prep'] = ''
df_retry2['url_prep'] = ''
for idx in df_retry2.index:
    temp_url = ''
    for col in df_retry2.columns[:5] :
        temp = df_retry2.loc[idx, col]
        if temp != '' :
            try:
                df_retry2.loc[idx, 'date_prep'] = int(temp)
            except:
                temp_url += df_retry2.loc[idx, col]
    df_retry2.loc[idx, 'url_prep'] = temp_url
df_retry2
```

```python
df_retry2_prep = df_retry2[['url_prep' ,'date_prep']]
df_retry2_prep.columns = ['url', 'date']
df_retry2_prep
```

```python
df_retry1.columns = ['url', 'date']
df_retry1
```

```python
df_retry_total = pd.concat([df_retry1, df_retry2_prep], axis=0)
df_retry_total
```

```python
df_retry_total = df_retry_total.reset_index(drop=True)
df_retry_total
```

```python
df_retry_total.to_csv('retry_url_total.csv', index=False)
```

```python
df_news_total = pd.concat([df_news_7y, df_news_10y, df_news_1y], axis=0)
df_news_total = df_news_total.reset_index(drop=True)
df_news_total
```

```python
drop_list = []
for idx in df_news_total.index :
    temp_date = df_news_total.loc[idx, 'date']
    if temp_date == 'date' :
        print(idx)
        drop_list.append(idx)
print(drop_list)
df_news_total = df_news_total.drop(drop_list)
df_news_total
```

```python
df_news_total['date'] = df_news_total['date'].astype('int')
df_news_total = df_news_total.drop_duplicates(subset=['url'])
df_news_total = df_news_total.drop_duplicates(subset=['content'])
df_news_total = df_news_total.sort_values(by=['date', 'media'])
df_news_total = df_news_total.reset_index(drop=True)
df_news_total
```

```python
df_news_total['content_len'] = df_news_total['content'].apply(len)
df_news_total
```

```python
df_news_total_up150 = df_news_total[df_news_total['content_len'] > 150]
df_news_total_up150 = df_news_total_up150.reset_index(drop=True)
df_news_total_up150.to_csv('naver_news_scrapy_total_prep.csv', index=False)
df_news_total_up150
```
